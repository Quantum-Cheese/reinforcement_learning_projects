### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

Note that **in this coding environment, you will not be able to watch the agent while it is training**, and you should set `train_mode=True` to restart the environment.

````python
env_info = env.reset(train_mode=True)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````

* define a function to plot the reward scores

````python
def plot_scores(scores,filename):
    # plot the scores
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(np.arange(len(scores)), scores)
    plt.ylabel('Score')
    plt.xlabel('Episode #')
    plt.savefig(filename)
    plt.show()
````

* define a fucntion to train an dqn agent and get the trainning scores  

I use the double DQN architecture for this project

````python
def train_dqn(agent,n_episodes=20000, max_t=5000,
        eps_start=1.0, eps_end=0.01, eps_decay=0.995):
    """"Params
    ======
        n_episodes (int): maximum number of training episodes
        max_t (int): maximum number of timesteps per episode
        eps_start (float): starting value of epsilon, for epsilon-greedy action selection
        eps_end (float): minimum value of epsilon
        eps_decay (float): multiplicative factor (per episode) for decreasing epsilon
    """
    scores = []  # list containing scores from each episode
    scores_window = deque(maxlen=100)  # last 100 scores
    eps = eps_start  # initialize epsilon

    start_time=time.time()
    for i_episode in range(1, n_episodes + 1):
        env_info = env.reset(train_mode=True)[brain_name] # reset the environment
        state = env_info.vector_observations[0]
        score = 0
        for t in range(max_t):
            # take an action using the local Q net
            action = agent.act(state, eps)
            try:
                
                # interact with the environment and get (s,r,a,s')
                env_info = env.step(action)[brain_name]        # send the action to the environment
                next_state = env_info.vector_observations[0]   # get the next state
                reward = env_info.rewards[0]                   # get the reward
                done = env_info.local_done[0]
            
                # transfer the (s,r,a,s') to dqn agent
                agent.step(state, action, reward, next_state, done)
                state = next_state
                score += reward
                if done:
                    break
            except Exception as e:
                pass
            
        scores_window.append(score)  # save most recent score
        scores.append(score)  # save most recent score
        eps = max(eps_end, eps_decay * eps)  # decrease epsilon with time step
        
        print('\rEpisode {}\t Average Score: {:.2f}'.format(i_episode,np.mean(scores_window)),end="")
        if i_episode % 100 == 0:
            print('\rEpisode {}\t Average Score: {:.2f}'.format(i_episode,np.mean(scores_window)))
            time_spend=time.localtime(time.time()-start_time)
            print("Running time:{}\n".format(time.strftime("%H:%M:%S",time_spend)))
        if np.mean(scores_window) >= 15.0:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode,
                                                                                         np.mean(scores_window)))
            torch.save(agent.qnetwork_local.state_dict(), 'dqn_cpu.pth')
            
            break
    time_spend=time.localtime(time.time()-start_time)
    print("\nTotal running time:{}".format(time.strftime("%H:%M:%S",time_spend)))
    return scores
    
````

#### Train  the Double DQN agent and plot the average scores

````python
from agent_dqn import AgentV2

agent=AgentV2(state_size=37, action_size=4, seed=0) 

# train the agent the get the scores
scores = train_dqn(agent) 
````

* define a function to test an trained agentï¼Œget the average reward scores for a given running time

````python
def trained_agent_test(agent,filename,episode_num=500,max_t=5000):
    agent.qnetwork_local.load_state_dict(torch.load(filename))
    scores=[]
    scores_window = deque(maxlen=100)
    for i_episode in range(1,episode_num+1):
        env_info = env.reset(train_mode=True)[brain_name] # reset the environment
        state = env_info.vector_observations[0]
        score = 0
        for t in range(max_t):
            # using greedy policy to choose an action at given state
            action = agent.act(state)
            
            # interact with the environment
            env_info = env.step(action)[brain_name]        
            next_state = env_info.vector_observations[0]   
            reward = env_info.rewards[0]                   
            done = env_info.local_done[0]
            state = next_state
            score += reward
            if done:
                break
        scores.append(score)
        scores_window.append(score)
        print('\rEpisode {}\t  Average Score: {:.2f}'.format(i_episode, np.mean(scores_window)),end="")
        if i_episode % 100 == 0:
            print('\rEpisode {}\t  Average Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
    return scores
            
````

#### Test an trained agent

````python
agent = AgentV2(state_size=37, action_size=4, seed=0)
test_scores=trained_agent_test(agent,'dqn_model.pth')
````
